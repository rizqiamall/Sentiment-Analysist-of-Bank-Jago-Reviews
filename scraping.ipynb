```python
from google_play_scraper import Sort, reviews
import pandas as pd

app_id = 'com.jago.digitalBanking'  # Ganti dengan ID aplikasi yang kamu teliti

result, _ = reviews(
    app_id,
    lang='id',  # Bahasa Indonesia
    country='id',
    sort=Sort.NEWEST,
    count=10000, 
    filter_score_with=None  # Bisa filter skor 1-5, kalau None ambil semua
)

df = pd.DataFrame(result)
df.to_csv("jago_reviews.csv", index=False)
print("Total review:", len(df))
print(df.head())
```
